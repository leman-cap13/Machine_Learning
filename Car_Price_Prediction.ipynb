```python
df['CarName']=df['CarName'].replace(df.groupby('CarName')['price'].mean().to_dict())
```

```python
def remove_outlier(df):
  num_columns=df.select_dtypes(include=np.number).columns
  for column in num_columns:
    Q3=df[column].quantile(0.25)
    Q1=df[column].quantile(0.75)
    IQR=Q3-Q1
    upper_bound=Q3+1.5*IQR
    lower_bound=Q1-1.5*IQR
    df=df[(df[column]>=upper_bound) & ( df[column]<=lower_bound)]
  return df
```

```python
df1=remove_outlier(df)
```

```python
df1
```

```python
X=df1.drop('price', axis=1)
y=df1['price'].copy()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2, random_state=13)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder,PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import  LinearRegression
```

```python
num_pipeline=Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('poly', PolynomialFeatures(degree=1)),
    ('encode', StandardScaler())
])
cat_pipeline=Pipeline([
    ('impute', SimpleImputer(strategy='constant', fill_value='missing')),
    ('encode', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

transformer=ColumnTransformer([
    ('num', num_pipeline, num_feature),
    ('cat', cat_pipeline, cat_feature)
],remainder='passthrough')

estimator=LinearRegression()

full_pipeline=Pipeline([
  ('preprocessing', transformer),
  ('estimator', estimator)

])
```

```python
full_pipeline.fit(X_train, y_train)
```

```python
full_pipeline.score(X_test,y_test)
```

```python
full_pipeline.score(X_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha=10.0)
rdg_pipeline=Pipeline([
  ('preprocessing', transformer),
  ('estimator', ridge_model)

])
```

```python
rdg_pipeline.fit(X_train,y_train)
```

```python
rdg_pipeline.score(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


predictions = full_pipeline.predict(X_test)
residuals = y_test - predictions


plt.figure(figsize=(10,6))
plt.scatter(predictions, residuals)
plt.hlines(y=0, xmin=min(predictions), xmax=max(predictions), colors='red')
plt.title("Residuals vs Predicted Values")
plt.xlabel("Predicted Car Price")
plt.ylabel("Residuals")
plt.show()
```

```python
from statsmodels.stats.stattools import durbin_watson


dw_stat = durbin_watson(residuals)
print('Durbin-Watson stat:', dw_stat)
```

```python
import scipy.stats as stats


plt.figure(figsize=(10,6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Q-Q Plot of Residuals")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error


mse = mean_squared_error(y_test, predictions)

print("Mean Squared Error (MSE):", mse)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error


mse = mean_squared_error(y_test, predictions)
print(f"Mean Squared Error (MSE): {mse}")


rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

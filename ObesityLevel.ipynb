```python
X=df.drop('NObeyesdad',axis=1)
y=df['NObeyesdad'].copy()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2,stratify=y,random_state=13)
```

```python
y_train
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_train
```

```python
y_test=encoder.fit_transform(y_test)
y_test
```

```python
num_feature=X_train.select_dtypes(include=np.number).columns
cat_feature=X_train.select_dtypes(exclude=np.number).columns
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
```

```python
num_pipeline =Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('encoder',StandardScaler())
])

cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='constant',fill_value='missing')),
    ('encoder',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

transformer=ColumnTransformer([
    ('num',num_pipeline,num_feature),
    ('cat',cat_pipeline,cat_feature)
],remainder='passthrough')

estimator=RandomForestClassifier()

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('estimator',estimator)
])
```

```python
full_pipeline.fit(X_train,y_train)
```

```python
full_pipeline.score(X_test,y_test)
```

```python
full_pipeline.predict(X_test.iloc[:5])
```

```python
y_test[:5]
```

```python
X_train_prepared=pd.DataFrame(transformer.fit_transform(X_train),columns=transformer.get_feature_names_out())
X_test_prepared=pd.DataFrame(transformer.transform(X_test),columns=transformer.get_feature_names_out())
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train_prepared,y_train)
rfc.score(X_test_prepared,y_test)
```

```python
sorted(zip(rfc.feature_importances_,X_train_prepared.columns))
```

```python
from sklearn.model_selection import GridSearchCV
params={
    'estimator__n_estimators': [70,100,130],
    'estimator__max_features': [7,9,19]
}

gscv=GridSearchCV(full_pipeline,params,cv=3,scoring='neg_mean_absolute_error')
gscv.fit(X_train,y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_estimator_
```

```python
gscv.best_estimator_.score(X_test,y_test)
```

```python
gscv.best_estimator_.predict(X_test.iloc[:5])
```

```python
from sklearn.model_selection import cross_val_predict

y_train_prediction = cross_val_predict(full_pipeline,X_train,y_train,cv=3)
y_train_prediction
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_train,y_train_prediction)
cm
```

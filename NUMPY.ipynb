```python
import pandas as pd

data = [[3, 108939], [2, 12747], [8, 87709], [6, 91796]]
accounts = pd.DataFrame(data, columns=['account_id', 'income']).astype({'account_id':'Int64', 'income':'Int64'})
```

```python
categories = pd.DataFrame({'category': ['High Salary', 'Low Salary', 'Average Salary']})
accounts['category']=accounts['income'].map(lambda x: 'Low Salary' if x < 20000 else ('Average Salary' if  20000 < x <50000 else 'High Salary'))
accounts = pd.merge(categories, accounts, on = 'category', how = 'left')
accounts.groupby('category').count().reset_index()[['category',	'account_id']].rename(columns = {'account_id': "accounts_count"})
# accounts.groupby('category').size().to_frame('accounts_count').reset_index()
```

```python
data = [[1, 'Leetcode Solutions', 'Book'], [2, 'Jewels of Stringology', 'Book'], [3, 'HP', 'Laptop'], [4, 'Lenovo', 'Laptop'], [5, 'Leetcode Kit', 'T-shirt']]
products = pd.DataFrame(data, columns=['product_id', 'product_name', 'product_category']).astype({'product_id':'Int64', 'product_name':'object', 'product_category':'object'})
data = [[1, '2020-02-05', 60], [1, '2020-02-10', 70], [2, '2020-01-18', 30], [2, '2020-02-11', 80], [3, '2020-02-17', 2], [3, '2020-02-24', 3], [4, '2020-03-01', 20], [4, '2020-03-04', 30], [4, '2020-03-04', 60], [5, '2020-02-25', 50], [5, '2020-02-27', 50], [5, '2020-03-01', 50]]
orders = pd.DataFrame(data, columns=['product_id', 'order_date', 'unit']).astype({'product_id':'Int64', 'order_date':'datetime64[ns]', 'unit':'Int64'})
orders
```

```python
products
```

```python
df1=orders[(pd.to_datetime(orders['order_date']).dt.month.reset_index()==2) & (pd.to_datetime(orders['order_date']).dt.year.reset_index()==2020)].dropna(subset='order_date')[['order_date']]
df1
```

```python
orders = orders[(orders['order_date'].dt.month == 2) & (orders['order_date'].dt.year == 2020)].groupby(['product_id'])['unit'].sum().reset_index()
orders
```

```python
orders = orders[orders['unit'] >= 100]
pd.merge(products, orders, on = 'product_id')[['product_name', 'unit']]
```

```python
val=orders.merge(df1,on='order_date',how='right')
val
```

```python
val1=val.groupby('product_id').sum('unit').reset_index()
val1=val1[val1['unit']>=100]
val1
```

```python
products.merge(val1,on='product_id')[['product_name','unit']]
```

```python
df=df.drop(['Month','Name','ID','Customer_ID','SSN','Type_of_Loan','Monthly_Inhand_Salary'],axis=1)
df
```

```python
df['Age'] = df['Age'].replace('_', '', regex = True)
df['Age'] = df['Age'].astype(int)
df = df[(df['Age'] <= 101) & (df['Age'] >= 18)]
```

```python
df.loc[:,'Annual_Income'] = df['Annual_Income'].replace('_', '', regex = True)
df.loc[:,'Annual_Income'] = df['Annual_Income'].astype(float)
df['Annual_Income']
```

```python
df.loc[:,'Num_of_Loan']=df['Num_of_Loan'].replace('_', '', regex = True)
df.loc[:,'Num_of_Loan']=df['Num_of_Loan'].astype(int)
df['Num_of_Loan']
```

```python
df['Credit_History_Age']=df['Credit_History_Age'].str.replace('Years', '').str.replace('Months', '').str.replace('and', '').str.strip()
#str.replace(r'\s+', '') remove aa whitespace
# df['Credit_History_Age']=df['Credit_History_Age'].str.replace(r'\s+', '')
df['Credit_History_Age']=df['Credit_History_Age'].str.replace(' ','')

df['Credit_History_Age']
```

```python
df.loc[:,'Payment_Behaviour']=df['Payment_Behaviour'].str.replace('_', ' ',regex=True)
df.loc[:,'Payment_Behaviour']=df['Payment_Behaviour'].str.strip('!@9#%8')
```

```python
df.loc[:, 'Outstanding_Debt']=df['Outstanding_Debt'].replace('_','',regex=True)
df['Outstanding_Debt']=df['Outstanding_Debt'].astype(float)
```

```python
df.loc[:,'Amount_invested_monthly']=df['Amount_invested_monthly'].replace('_','',regex=True)
df.loc[:,'Amount_invested_monthly']=df['Amount_invested_monthly'].astype(float)
df.loc[:,'Amount_invested_monthly']=df['Amount_invested_monthly'].fillna(df['Amount_invested_monthly'].median())
```

```python
df.loc[:,'Monthly_Balance']=df['Monthly_Balance'].replace('_','',regex=True)
df['Monthly_Balance']=df['Monthly_Balance'].astype(float)
```

```python
df.loc[:,'Monthly_Balance']=df['Monthly_Balance'].fillna(df['Monthly_Balance'].median())
df['Monthly_Balance'].isna().sum()
```

```python
df.loc[:, 'Num_Credit_Inquiries']=df['Num_Credit_Inquiries'].fillna(df['Num_Credit_Inquiries'].median())
df['Num_Credit_Inquiries']
```

```python
df.loc[:,'Changed_Credit_Limit']=df['Changed_Credit_Limit'].replace('_','',regex=True)
df['Changed_Credit_Limit'] = pd.to_numeric(df['Changed_Credit_Limit'], errors='coerce')
# df['Changed_Credit_Limit'].str.contains(r'[^a-zA-Z0-9]').sum()
df['Changed_Credit_Limit'].astype(float)
```

```python
num_feature= X_train.select_dtypes(include=[np.number]).columns
cat_feature= X_train.select_dtypes(exclude=[np.number]).columns
```

# we have outlier in Interest_Rate

```python
Q1 = df['Interest_Rate'].quantile(0.25)
Q3 = df['Interest_Rate'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (df['Interest_Rate'] < lower_bound) | (df['Interest_Rate'] > upper_bound)
df = df[~outliers]
```

```python
cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='constant',fill_value='missing')),
    ('onehotencoder',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

num_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer = ColumnTransformer([
    ('num',num_pipeline,num_feature),
    ('cat',cat_pipeline,cat_feature)

],remainder='passthrough')

estimator=RandomForestClassifier()

full_pipeline = Pipeline([
    ('preprocessing', transformer),
    ('estimator',  estimator)

])


```

```python
pd.DataFrame(transformer.fit_transform(X_train,y_train), columns = transformer.get_feature_names_out())
```

```python
full_pipeline.fit(X_train,y_train)
```

```python
full_pipeline.score(X_test,y_test)
```

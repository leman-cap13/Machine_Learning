```python
X=df.drop('price_range',axis=1)
y=df['price_range'].copy()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
num_feature=X_train.select_dtypes(include=[np.number]).columns
cat_feature=X_train.select_dtypes(exclude=[np.number]).columns
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing  import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
num_pipeline=Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])

#Useless code
# cat_pipeline = Pipeline([
#     ('impute', SimpleImputer(strategy='constant',fill_value='missing')),
#     ('onehotencoder',OneHotEncoder(handle_unknown='ignore',sparse_output='ignore'))
# ])

transformer=ColumnTransformer([
    ('num',num_pipeline, num_feature)
    # ('cat',cat_pipeline,cat_feature)
],remainder='passthrough')

estimator= RandomForestClassifier()

full_pipeline =Pipeline([
    ('preprocessing', transformer),
    ('estimator',estimator)
])
```

```python
full_pipeline.fit(X_train,y_train)
```

```python
full_pipeline.score(X_test, y_test)
```

```python
full_pipeline.predict(X_test.iloc[:10])
```

```python
y_test.iloc[:10]
```

```python
X_train_prepared=pd.DataFrame(transformer.fit_transform(X_train),columns=transformer.get_feature_names_out())
X_test_prepared=pd.DataFrame(transformer.transform(X_test),columns=transformer.get_feature_names_out())
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()

rfc.fit(X_train_prepared, y_train)
rfc.score(X_test_prepared, y_test)
```

```python
sorted(zip(rfc.feature_importances_, X_train_prepared.columns))

```

```python
from sklearn.model_selection import GridSearchCV # RandomizedSearchCV
```

```python
param_grid = {
    'estimator__n_estimators': [175, 176, 177, 178, 179]
}

gscv = GridSearchCV(full_pipeline, param_grid, cv=3)
gscv.fit(X_train, y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_estimator_.fit(X_train, y_train)
```

```python
gscv.best_estimator_.score(X_train, y_train), gscv.best_estimator_.score(X_test, y_test)
```
